Investigate the variables

```python
# passenger_count =0 & total_amount !=0

df[(df['passenger_count']==0) & (df['total_amount']>=0)]
```

The most expensive rides are not necessarily the longest ones.

```python
df[['trip_distance', 'total_amount']].sort_values(by=['trip_distance'],ascending=False).head(10)
```

```python
# payement method & tip_amount

df.groupby('payment_type')['tip_amount'].agg(['count', 'mean'])
```

```python
#vendor_id wrt total_amount & tip
df.groupby('VendorID')['total_amount', 'tip_amount'].agg(['mean','sum'])
```

```python
# vendor_id, mean is same but total_amount is different- check distance

df.groupby('VendorID')['total_amount', 'trip_distance'].agg(['mean','sum'])
```

```python
# credit card and passengers

credit_card = df[df['payment_type']==1]
credit_card['passenger_count'].value_counts()
```

```python
# passengers & tip amount
df.groupby('passenger_count').mean(numeric_only = True)[['tip_amount']]
```

```python
# weekdays relation with tip_amount, total_amount, trip_distance
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['trip_weekday'] = df['tpep_pickup_datetime'].dt.day_name()

df.groupby('trip_weekday')[['tip_amount', 'total_amount', 'trip_distance']].mean()
```

```python
# explore DOLocationID

df.groupby(['DOLocationID', 'PULocationID'])['VendorID'].count()
```

```python
# 216 unique location
df['DOLocationID'].value_counts()
```

```python
# 152 unique locations
df['PULocationID'].value_counts()
```

**Conclusion:**
After looking at the dataset, the two variables that are most likely to help build a predictive model for taxi ride fares are total_amount and trip_distance because those variables show a picture of a taxi cab ride.
