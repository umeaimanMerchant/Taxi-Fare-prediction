**The goal** is to apply descriptive statistics and hypothesis testing in Python. <h3>Imports and data loading</h3>

```python
# import library

import pandas as pd
import scipy.stats as stats
```

```python
# import dataset

taxi_data = pd.read_csv("2017_Yellow_Taxi_Trip_Data.csv", index_col = 0)
```

<h3>Conduct EDA and hypothesis testing</h3>

```python
taxi_data.head()
```

```python
taxi_data.describe(include = 'all')
```

```python
taxi_data.groupby('payment_type')['fare_amount'].mean()
```

**Hypothesis Testing:-**

Null hypothesis: There is no difference in average fare between customers who use credit cards and dispute. 

Alternative hypothesis: There is a difference in average fare between customers who use credit cards and dispute

```python
significance_level = 0.05

credit_card = taxi_data[taxi_data['payment_type'] == 1]['fare_amount']
cash = taxi_data[taxi_data['payment_type'] == 4]['fare_amount']

stats.ttest_ind( a = credit_card, b = cash, equal_var= False)
```

**Conclusion-**

p_value > significance_level

Fail to reject null hypothesis

Null hypothesis is true: There is no difference in average fare between customers who use credit cards and dispute.

**Hypothesis Testing:** 
**Null hypothesis**: There is no difference in average fare between customers who use credit cards and customers who use cash. 
**Alternative hypothesis**: There is a difference in average fare between customers who use credit cards and customers who use cash

```python
# other payment types
credit_card = taxi_data[taxi_data['payment_type'] == 1]['fare_amount']
other = taxi_data[taxi_data['payment_type'] == 2]['fare_amount']

stats.ttest_ind( a = credit_card, b = other, equal_var= False)
```

**Conclusion-**

p_value << significance_level

Reject null hypothesis

Alternative hypothesis is true: There is a difference in average fare between customers who use credit cards and customers who use cash
